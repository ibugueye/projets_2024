```python
df = pd.read_csv("df_final.csv")
```

```python
# Splitting data into train and test
X = df.drop('TARGET', axis=1)
y = df['TARGET']
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify =y, random_state=0, train_size = 0.10)
```

```python
import shap
import xgboost as xgb

# Entranement d'un modle XGBoost (ou tout autre modle)
model = xgb.XGBClassifier().fit(X_train, y_train)

# Cration de l'expliqueur SHAP et calcul des valeurs SHAP
explainer = shap.Explainer(model)
shap_values = explainer(X_test)

# SHAP Summary Plot pour une interprtation globale
shap.summary_plot(shap_values, X_test)

# SHAP Waterfall Plot pour une interprtation locale d'une instance spcifique
shap.plots.waterfall(shap_values[0])
```

```python

import shap
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Entranement d'un modle RandomForest
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Cration de l'expliqueur SHAP et calcul des valeurs SHAP
explainer = shap.Explainer(model)
shap_values = explainer(X_test)

# SHAP Summary Plot pour une interprtation globale
shap.summary_plot(shap_values, X_test, show=False)
plt.show()

# SHAP Waterfall Plot pour une interprtation locale d'une instance spcifique
shap.plots.waterfall(shap_values[0])
plt.show()
```
